```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("CO2 Emissions_Canada.csv")
data.columns
x = data["Engine Size(L)"]
y = data["CO2 Emissions(g/km)"]
n = float(len(x))
m= 0
c = 0
l = 0.0001
itr = 1000
for i in range(itr):
    y_p = m*x + c
    Dm = (-2/n)*sum(x*(y- y_p))
    Dc = (-2/n)*sum(y-y_p)
    m = m - l*Dm
    c = c - l*Dc
print(m,c)
```

```python
y_p = m*x + c
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_p),max(y_p)],color = "red")
```

```python
def predict(inpt,m,c):
    value = m*inpt + c
    return value
engine_size = 3
a = predict(engine_size,m,c)
print(a)
```
